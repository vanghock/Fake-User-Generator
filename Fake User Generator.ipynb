```python
from faker import Faker
from datetime import datetime, timedelta
import random
from tabulate import tabulate

fake = Faker()

# List of USA cities
usa_cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose']

def generate_user_data(n):
    user_data = []
    for _ in range(n):
        # Generate woman's name and surname
        name = fake.first_name_female()
        surname = fake.last_name_female()

        # Generate account creation date within the last 3 months
        end_date = datetime.now()
        start_date = end_date - timedelta(days=90)
        account_date = fake.date_between(start_date, end_date)

        # Generate age (let's assume it's between 18 and 80 for this case)
        age = random.randint(18, 60)

        # Generate email
        email = fake.email()

        # Select city from USA
        city = random.choice(usa_cities)

        full_name = f"{name} {surname}"  # Combine name and surname

        user_data.append((full_name, account_date, age, email, city))
    return user_data

# Generate user data
data = generate_user_data(15)

# Convert the data to a list of lists for tabulate
table_data = [list(user) for user in data]

# Define the table headers
headers = ['Name', 'Account Creation Date', 'Age', 'Email', 'City']

# Print the table
print(tabulate(table_data, headers, tablefmt="grid"))


```
